```r
library("imputeLCMD");
```

```r
obs_1 <- read.table("~/dream_data/data_obs_1.txt", sep = '\t', fill = T, header = T, row.names = 1)
obs_1_imputed <- impute.MinDet(obs_1, q = 0.01)
```

```r
head(obs_1)
```

```r
head(obs_1_imputed)
```

```r
truth <- read.table("~/dream_data/data_true.txt", sep = '\t', fill = T, header = T, row.names = 1)
```

```r
head(truth)
```

```r
s1_true = truth$Sample_1
s1_obs_1 = obs_1$Sample_1
s1_missing = s1_true[is.na(s1_obs_1)]
```

```r
head(s1_missing)
```

```r
s1_impute = obs_1_imputed$Sample_1
s1_missing_impute = s1_impute[is.na(s1_obs_1)]
head(s1_missing_impute)
```

```r
hist(s1_missing, breaks = 50)
abline(v = s1_missing_impute[1], col = "blue", lwd = 2)
```

```r
for (sample in c("Sample_5", "Sample_2", "Sample_3")) {
    s_true = truth[sample]
    s_obs_1 = obs_1[sample]
    s_missing = s_true[is.na(s_obs_1)]
    s_impute = obs_1_imputed[sample]
    s_missing_impute = s_impute[is.na(s_obs_1)]
    hist(s_missing, main = sample, breaks = 50)
    abline(v = s_missing_impute[1], col = "blue", lwd = 2)
}
```
